# Bar Graph Exercise

Creating bar graphs using a dataset from an Alex The Analyst **[video](https://youtu.be/JpSMse3eVVg)** about creating visulizations in pandas. 

```python
# importing packages
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#reading in CSV

df = pd.read_csv(r"/Users/scipio/Downloads/Ice Cream Ratings.csv")

#setting index

df = df.set_index('Date')

#creating a variable for colors 

random = ['green', 'yellow','red']
```

```python
# creating subplot dimensions

fig,axs = plt.subplots(nrows =1, ncols =3, figsize = (15,5))

# creating bar garphs to insert into subpplot figure
# insert bar graph into subplot via indexing axs in subplot tuple fig,axs above

df['Overall Rating'].plot(kind = 'bar', grid = True, ylabel = 'Value', title = 'Overall Rating', ax = axs[2], color = 'orange')
df['Flavor Rating'].plot(kind = 'bar', grid = True, ylabel = 'Value', title = 'Flavor Rating', ax = axs[0], color = 'blue')
df['Texture Rating'].plot(kind = 'bar', grid = True, ylabel = 'Value', title = 'Texture Rating', ax = axs [1], color = 'green')


plt.show()
```

```python
# all values in the same bar graph 

#list of colors 

default = ['blue', 'green', 'orange']

df.plot(kind = 'bar', color = default, title = 'Ice Cream Ratings', figsize = (10,5), grid = True, ylabel = 'Value')
```
